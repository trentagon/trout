# Plot Methods

This notebook does the calculations and plots the results from the methods portion of the paper.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Figure 2: Water content of magma

```python
# Calculate water in magma
WMF = np.linspace(0.01,12,500)/100
D = 0.001

F30 = 0.3
m_melt30 = (WMF/F30)*(1-np.power((1-F30),1/D))

F20 = 0.2
m_melt20 = (WMF/F20)*(1-np.power((1-F20),1/D))

F15 = 0.15
m_melt15 = (WMF/F15)*(1-np.power((1-F15),1/D))

F10 = 0.1
m_melt10 = (WMF/F10)*(1-np.power((1-F10),1/D))

gpa1 = 1
m_sat1 = (12*np.power(gpa1,0.6)) + (1*gpa1)

gpa2 = 2
m_sat2 = (12*np.power(gpa2,0.6)) + (1*gpa2)

gpa3 = 3
m_sat3 = (12*np.power(gpa3,0.6)) + (1*gpa3)

gpa5 = 5
m_sat5 = (12*np.power(gpa5,0.6)) + (1*gpa5)
```

```python
plt.rcParams['font.family'] = 'Gill Sans'
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['font.size'] =  12

# col_dots = "#172713"
col_dots = 'k'
col_earth = "#009465"
col_mars = "#cb2f43"
col_io = "#59256a"
```

```python
lwdt = 1
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(8,4)

ax.minorticks_on()

idx_cut1 = 110
ax.fill_between(WMF[0:idx_cut1]*100,m_melt20[0:idx_cut1]*100,m_melt10[0:idx_cut1]*100, color = col_dots,alpha = 0.25,edgecolor='none')

idx_cut2 = 109
idx_cut3 = 126
ax.fill_between(WMF[idx_cut2:idx_cut3]*100,m_sat1*np.ones_like(WMF[idx_cut2:idx_cut3]),m_sat3*np.ones_like(WMF[idx_cut2:idx_cut3]), 
                color = col_dots,alpha = 0.25,edgecolor='none')

l_fs = 12


lwdt2 = 1
alp = 1
ax.axhline(m_sat1,linestyle='--',label='Water solubility limit',color=col_dots,alpha=alp,linewidth=lwdt2)
# ax.axhline(m_sat2,linestyle='--',color='grey',alpha=alp,linewidth=lwdt2)
ax.axhline(m_sat3,linestyle='--',color=col_dots,alpha=alp,linewidth=lwdt2)
# ax.axhline(m_sat5,linestyle='--',color='grey',alpha=alp,linewidth=lwdt2)

ax.plot(WMF*100,m_melt10*100,'-',label='Water abundance limit',linewidth=lwdt,color=col_dots)
# ax.plot(WMF*100,m_melt15*100,'-',linewidth=lwdt,color='black')
ax.plot(WMF*100,m_melt20*100,'-',linewidth=lwdt,color=col_dots)

ax.text(0.5,m_sat1+0.5,'$P_{prim} = 1$ GPa',fontsize=l_fs,color=col_dots)
# ax.text(10.4,m_sat2+0.5,'$P_{prim} = 2$ GPa',c='grey',fontsize=l_fs)
ax.text(0.5,m_sat3+0.5,'$P_{prim} = 3$ GPa',fontsize=l_fs,color=col_dots)
# ax.text(10.4,m_sat5+0.5,'$P_{prim} = 5$ GPa',c='grey',fontsize=l_fs)

ax.text(1.7,18,'F = 10%',fontsize=l_fs,rotation=21,color=col_dots)
# ax.text(4.25,30,'F = 15%',c='black',fontsize=l_fs,rotation=45)
ax.text(1.7,6.5,'F = 20%',fontsize=l_fs,rotation=10,color=col_dots)

# Draw a box on the plot
rect = plt.Rectangle((0.18, 0.1), 0.54 - 0.18, 1.5 - 0.1, linewidth=1, edgecolor=col_earth, facecolor=col_earth)
ax.add_patch(rect)

# Annotate the box
ax.text(0.55, 0.1, 'MORB', fontsize=l_fs, color=col_earth)

ax.set_xlim(0,3)
ax.set_ylim(0,40)

ax.set_xlabel('$m_{H_2O}^{mantle}$ (wt%) \n Solid mantle water content',fontsize = l_fs)
ax.set_ylabel('$m_{H_2O}^{melt}$ (wt%) \n Magma water content', fontsize = l_fs)
ax.legend(ncol=1,columnspacing=.8,labelspacing=.4,fontsize=l_fs,loc='upper left')

plt.show()
```
